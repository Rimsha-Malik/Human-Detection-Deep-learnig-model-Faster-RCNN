```
#import keras
import os
import cv2 #open  cv (use for images like image readimg ,image shoe etc)
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import h5py
import sklearn
```

```
#reading train.csv and store it in pandas dataframe
trainRCNN = pd.read_csv('train.csv', sep=",", header=None)
#trainRCNN.columns = ['image_names','xmin' ,'ymin' , 'xmax','ymax','cell_type']
trainRCNN.columns = ['filename','cell_type','xmin','xmax','ymin','ymax']
trainRCNN.head()#put 10 in brackets will give 10 rows
```

```
#for annotate.txt
f= open('/content/drive/My Drive/keras-frcnn/keras-frcnn-master' + "/annotate.txt","w+")
for idx, row in trainRCNN.iterrows():

    x1 = int(row['xmin'] )
    x2 = int(row['xmax'] )
    y1 = int(row['ymin'] )
    y2 = int(row['ymax'] )
    fileName =row['filename']
    className = row['cell_type']
    f.write(fileName + ',' + className + ',' + str(x1) + ',' + str(x2)+ ',' + str(y1)  + ',' + str(y2) +'\n')
f.close()
```
